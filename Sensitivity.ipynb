# What is a good planar target for 3D camera calibration?

We don't need 3D calibration targets to calibrate a camera's pose and focal length. Planar calibration targets suffice. But the choice of a calibration target matters quite a bit. I show that a rectangular target is much better than a triangular one.

```python
%matplotlib inline

import pylab as P
from numpy import *
import pandas as pd

import sensitivity as S  # All the work lives here

# ground truth camera pose: 
# Rotations (wx,wy,wz) are a Cayley parametrization of rotation (see the code). It was easier to
# parametrize rotation that way than a screw, or Euler angles, and it does not over parametrize
# rotation like quaternions. (tx,ty,tz) are a translation, and f is the focal length
#    wx wy wz    tx  ty  tz    f
cam=(1.,0.,0.,   0.,-1.4,4.,  300.)
```

The package `sensitivty` contains a routine for absolute camera calibration. Given the coordinates of points in robot coordinates and their projections onto the image plane, it recovers a 7-dimension vector that represents the camera's pose and focal length relative to the robot coordinate system (3 parameters for rotation, 3 for translation, and one for focal length). This routine also requires an initial guess of the 7-dimensional pose parmeters.

I investigate this routine's sensitivity to different calibration targets and different starting guesses.

Here are two targets in robot coordinates and in image coordinates under a particular camera pose.

Some things to notice about both targets

1. Estimation error for all parameters grows linearly with observation noise.

1. Calibrating with the rectangular target is much more stable under observation noise.

1. Perturbing the initializer does not produce significantly worse answers: The colored dots are evenly mixed vertically.


# Conclusion

The triangle target, despite having the same size and nearly the same number of points as the rectangle target, is much more sensitive to observation noise. We should use a grid target instead.
